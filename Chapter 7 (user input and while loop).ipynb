```python
#try It yourself  7-5. Movie Tickets: 

prompt = "\nEnter your age, I will tell you your ticket cost. "
prompt += "\n Enter 'quit' when you are finished."

while True:
    
    ticket = input(prompt)
    
    if  ticket == 'quit':
         break

    ticket = int(ticket)
        
    if ticket < 3:
        cost = 'free'
            
    elif ticket >= 3 and ticket < 12:
        cost =  10 
    
    elif ticket >  12:
        cost =  15 
                
    print('Your ticket cost is '+str(cost)+'.')
       
  
```

```python
#try It yourself 7-8. Deli:


sandwich_orders = ['club','chicken','vegetable','cheese']
finished_sandwiches = []

while sandwich_orders:
        current_sandwich = sandwich_orders.pop()
        print("I' m making "+ current_sandwich+ ' sandwich.')
        finished_sandwiches.append(current_sandwich)
        
print('\nYou ordered these sandwiches:')
for finished_sandwiche in finished_sandwiches:
    print(finished_sandwiche.title())
```
